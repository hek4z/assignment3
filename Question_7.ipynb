## Part (4)

```python
import numpy as np
from sklearn.cluster import KMeans

k = 10 # telling us how many clusters we want in the data

# Create a KMeans model with k clusters
kmeans = KMeans(n_clusters=k, random_state=0)

# Fit the model to the data
kmeans.fit(x_test)

# Get cluster assignments for each data point
cluster_labels = kmeans.predict(x_test)
```

## Part (5)

```python
import pandas as pd
from sklearn.metrics import confusion_matrix

# Assuming you have the cluster assignments in 'cluster_labels' and the true labels in 'y_test'

# Create a DataFrame to hold the cluster assignments and true labels
df = pd.DataFrame({'Cluster': cluster_labels, 'TrueLabel': y_test})

# Create a confusion matrix to compare the cluster assignments with true labels
confusion = pd.crosstab(df['Cluster'], df['TrueLabel'])

# Print the confusion matrix to evaluate correspondence
print(confusion)

# Calculate the proportion of digits that are clustered correctly
correctly_clustered = np.sum(np.max(confusion, axis=1)) / len(y_test)
print(f"Proportion of correct clustering: {correctly_clustered:.2f}")
```

The correspondence in this table is not very good. The highest values in the table do not lie on this diagonal as it should be if the test had high accuracy. A little under 60% of the clusters are correct, only 59% to be exact. This proportion, which is barely over 55% demonstrates that kMeans clustering is not a reliable model for this situation. The digits that are the hardest to distinguish from one another are 0, which had 812 values in the 6 cluster, 6, which had 782 values in the 9 cluster, 3, which had 703 values in the 0 cluster, and 2, which had 692 values in the 8 cluster. From the data that I have collected based on these tests, it seems as though k MC cannot well-recover the latent digits 0 - 9 without even knowing what those digits were.

## Part (6)

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from keras.datasets import mnist

# Load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Combine the training and test images
combined_data = np.vstack((train_images, test_images))

# Flatten the images (reshape them into 1D arrays)
combined_data = combined_data.reshape(combined_data.shape[0], -1)

# Perform PCA on the flattened data
pca = PCA()
pca.fit(combined_data)

# Calculate the explained variance for each component
explained_variance = pca.explained_variance_ratio_

# Create a scree plot
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance)
plt.xlabel("Number of Principal Components")
plt.ylabel("Explained Variance")
plt.title("Scree Plot")
plt.xlim(0, 50)
x_ticks = [5,10,13,15,20,25,30,35,40,45,50]
plt.xticks(x_ticks)
plt.grid(True)
plt.show()
```

The scree plot does not seem to use 10 as the k point. The scree plot overall is very unclear about which k would be best, but to me it looks like 13 would be one of the best options. Overall, there is not a super clear 'elbow' point to indicate which digits it tends to combine into the same classification. If I had to guess, it looks like 13 and 14 have a pretty similar slope and might get confused.
